```python
import pandas as pd
import polars as pl
```

```python
data_list_pd = [[1,"Alice"],[2,"Bob"],[3,"Charlie"]]
```

```python
pd.DataFrame(data_list_pd)
```

If no column name is provided, Pandas will provide default column names like 0,1,..., which is not informative at all. So, we can name our columns using an additional argument "columns".

```python
pd.DataFrame(data_list_pd,columns=["id","name"])
```

Polars also has dataframe, but the "columns" argument is replaced with the name "schema" here.

```python
pl.DataFrame(data_list_pd,schema=["id","name"])
```

Note that Polars gives us some extra information like data size ("shape" at the top of the data table), data type of each column (immediately after the column name), which Pandas doesn't do.

# Dealing with missing data

```python
pd_na=pd.DataFrame({'A':[1,2,None],"B":[4,None,6]})
pl_na=pl.DataFrame({'A':[1,2,None],"B":[4,None,6]})
```

```python
pd_na
```

```python
pd_na.dropna() # to remove fields with missing data
```

```python
pl_na.drop_nulls() # similar to dropna in pandas
```

```python
pd_na.fillna(0) # to replace missing data with 0
```

```python
pl_na.fill_null(0) # similar to fillna
```

```python
pl_na.fill_nan(0) # fills the empty fields with "null"
```

```python
pd_na.interpolate() # to fill the missing field based on the existing data
```

Polars has nothing like interpolate, so conversion may be needed to implement this feature on a Polars dataset.

# Merging and Joining

```python
df1=pd.DataFrame({"key":["A","B","C"],"value1":[1,2,3]})
df2=pd.DataFrame({"key":["A","B","C"],"value2":[4,5,6]})
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
df3=pd.DataFrame({"key":["A","B"],"value3":[7,8]})
```

```python
pd.merge(df1,df3,on="key",how="left") # fill the empty fields based on the left argument
```

```python
pd.merge(df1,df3,on="key",how="right") # fill the empty fields based on the right argument
```

```python
pd.merge(df1,df3,on="key",how="inner")
```

```python
pd.merge(df1,df3,on="key",how="outer")
```

```python
df1.set_index("key").join(df3.set_index("key"),how="inner")
```

```python
pd.concat([df1,df3])
```

```python
pd.concat([df1,df3],axis=0)
```

```python
pd.concat([df1,df3],axis=1)
```
